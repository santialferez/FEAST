```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
from feast import feast
```

```python
selector = feast(3,criterion='CMIM')
```

```python
selector.fit(X,y)
```

```python
X.dtype
```

```python
from sklearn import preprocessing
import numpy as np

data = preprocessing.minmax_scale(X)

# discretize each feature
n_bins = 10 # number of bins by default
X_n_samples, X_n_features = data.shape
X_discrete = np.zeros((X_n_samples, X_n_features))
bins = np.linspace(0, 1, n_bins)
for i in range(X_n_features):
    X_discrete[:, i] = np.digitize(data[:, i], bins)
X_discrete = X_discrete-1

data = X_discrete
```

```python
data
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
pipe = make_pipeline(selector, LinearDiscriminantAnalysis())

cross_val_score(pipe,X,y,cv=5)
```
